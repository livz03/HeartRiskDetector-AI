### Load data, visualize, check nulls ,correlations 

```python
#import libraries
import numpy as pd 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
#Load dataset
heart_df = pd.read_csv('../Datasets/heart.csv')
mitbih_df =pd.read_csv('../Datasets/mitbih_test.csv')
```

```python
#Basic Info
print("Heart dataset shape:", heart_df.shape)
print("Mitbih dataset shape:", mitbih_df.shape)
print(heart_df.head())
print(mitbih_df.head())
print(heart_df.info())
print(mitbih_df.info())
print(heart_df.describe())
print(mitbih_df.describe())
```

```python
# check for missing values
print("Heart missing values:\n",heart_df.isnull().sum())
print("Mitbih missing values:\n",mitbih_df.isnull().sum())
```

```python
# Visualizations (Heart Dataset)
sns.countplot(x='ChestPainType', data=heart_df)
sns.countplot(y='Cholesterol', data=heart_df)
plt.title("Heart Disease Distribution")
plt.grid()
plt.show()
```

```python
# Correlation Heatmap (numeric columns only)
plt.figure(figsize=(12,8))
numeric_cols = heart_df.select_dtypes(include=['number'])
sns.heatmap(numeric_cols.corr(), annot=True, cmap="coolwarm")
plt.title("Heart Dataset Correlation Heatmap")
plt.grid()
plt.show()
```
