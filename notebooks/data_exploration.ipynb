```python
# Load Required Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
# Load Multiple Datasets (Auto from Folder)

# Set the folder path where your datasets are
folder_path = "Datasets"

# Get all CSV files from the folder
dataset_files = [f for f in os.listdir(folder_path) if f.endswith(".csv")]

# Load all datasets into a dictionary
datasets = {}
for file in dataset_files:
    file_path = os.path.join(folder_path, file)
    datasets[file] = pd.read_csv(file_path)
    print(f"âœ… Loaded: {file} | Shape: {datasets[file].shape}")
```

```python
# Example Plot(Loop over all datasets)

for name, df in datasets.items():
    if 'heart_rate' in df.columns:
        print(f"ðŸ“Š Plotting heart_rate for: {name}")
        sns.histplot(df['heart_rate'], kde=True)
        plt.title(f"Heart Rate Distribution - {name}")
        plt.xlabel("Heart Rate")
        plt.ylabel("Frequency")
        plt.show()
    else:
        print(f"âŒ 'heart_rate' column not found in: {name}")
```
